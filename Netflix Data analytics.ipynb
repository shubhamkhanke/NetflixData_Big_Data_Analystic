#Q.2 : In which year highest no. of the TV Shows & Movies were realeased ? show with Bar Graph.


```python
data.dtypes
```

```python
data['Date_N'] = pd.to_datetime(data['date_added'])
```

```python
data.dtypes
```

```python
data['Date_N'].dt.year.value_counts()          #It counts the occurance of all individual years in Date columns
```

```python
import matplotlib.pyplot as plt
data['Date_N'].dt.year.value_counts().plot(kind = 'bar' , color = 'g')

plt.xlabel("Years")
plt.ylabel("Counts")
plt.title("Highest no. of the TV Shows & Movies")
plt.show()
```
